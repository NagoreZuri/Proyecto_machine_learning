```python
# toca ahora trabajar las variables no numéricas .

# Eliminar nulos en variables clave
df_final = df_final.dropna(subset=['averageRating', 'numVotes', 'runtimeMinutes', 'startYear', 'genres', 'titleType'])
```

```python
df_final   #cnteo de géneros y cuand sean menos de X, guardarlo en una columna extra como otros, eliminand la original
```

```python
# Calcular la media de averageRating por actor (primaryName)
actor_mean_rating = df_final.groupby('primaryName')['averageRating'].mean()

# Mapear la media de cada actor a una nueva columna en df_final
df_final['actor_mean_rating'] = df_final['primaryName'].map(actor_mean_rating)

# Mostrar algunos valores únicos de actores y su media
df_actors = df_final[['primaryName', 'actor_mean_rating']].drop_duplicates()
df_actors['actor_mean_rating'] = df_actors['actor_mean_rating']
df_final['actor_mean_rating'] = df_final['actor_mean_rating'].round(1)
```

```python
# Veamos la distribución de ratings y de NumVotes

# Añadir columna logarítmica (evitar log(0) con +1)
df_final['log_numVotes'] = np.log10(df_final['numVotes'] + 1)

# Crear los histogramas
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Histograma de averageRating
sns.histplot(df_final['averageRating'], bins=20, kde=True, ax=axes[0], color='skyblue')
axes[0].set_title('Distribución de averageRating')
axes[0].set_xlabel('averageRating')

# Histograma de log(numVotes)
sns.histplot(df_final['log_numVotes'], bins=20, kde=True, ax=axes[1], color='salmon')
axes[1].set_title('Distribución de log(numVotes + 1)')
axes[1].set_xlabel('log10(numVotes + 1)')

plt.tight_layout()
plt.show()
```

usar la clumna logarítmica Permite ver mejor las diferencias entre películas con pocos votos y me Facilita la visualización 

```python
# Crear columna categórica para averageRating en df_final con nuevos rangos
bins = [0, 4.5, 6.5, 8.9, 10]
labels = ['mala', 'regular', 'buena', 'muy buena']
df_final['rating_cat'] = pd.cut(df_final['averageRating'], bins=bins, labels=labels, right=True, include_lowest=True)
df_final[['averageRating', 'rating_cat']]

# Reordenar las columnas para que 'rating_cat' esté justo después de 'averageRating'
cols = list(df_final.columns)
cols.insert(cols.index('averageRating') + 1, cols.pop(cols.index('rating_cat')))
df_final = df_final[cols]
```

```python
df_final.drop(columns=['directors', 'nconst'], inplace=True)
```

```python
df_final
```

```python
#  vamos a solucionar el tema de los géneros ya que hay demasiados y muy variados 
#  Separar géneros por fila y obtener lista única
df_final['genres_list'] = df_final['genres'].fillna('').apply(lambda x: x.split(','))

# Explode para contar frecuencia de cada género
all_genres = df_final.explode('genres_list')['genres_list'].value_counts()

# Establecer umbral para decidir si un género es poco frecuente (< 100 apariciones)
rare_genres = all_genres[all_genres < 100].index.tolist()
common_genres = all_genres[all_genres >= 100].index.tolist()

#  Crear columnas binarias SOLO para géneros frecuentes
for genre in common_genres:
    df_final[genre] = df_final['genres_list'].apply(lambda x: 1 if genre in x else 0)

# Crear columna 'other_genres' binaria si hay al menos un género raro en esa fila
df_final['other_genres'] = df_final['genres_list'].apply(lambda x: 1 if any(g in rare_genres for g in x) else 0)

#  Eliminar columna auxiliar
df_final.drop(columns=['genres_list'], inplace=True)
```
