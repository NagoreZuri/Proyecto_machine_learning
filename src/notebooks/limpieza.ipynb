```python

df_titulo_rating = pd.read_csv('C:/Users/nagor/Documents/Thebridge/Machine_Learning/imdb/datasets_brutos/title.ratings.tsv.gz', 
                 sep='\t',          # separador de tabulaciones
                 compression='gzip',
                 low_memory=False)  #
```

```python
df_titulo_rating
```

```python
chunks = pd.read_csv(
    'C:/Users/nagor/Documents/Thebridge/Machine_Learning/imdb/datasets_brutos/title.principals.tsv.gz',
    sep='\t',
    compression='gzip',
    chunksize=1_000_000,  # Carga un mill√≥n de filas a la vez
    low_memory=False
)

# Ejemplo: combinar todos los chunks en uno filtrando en el proceso
filtered_chunks = []
for chunk in chunks:
    filtered = chunk[chunk['category'] == 'actor']  # ejemplo de filtrado
    filtered_chunks.append(filtered)

df_principals = pd.concat(filtered_chunks)
```

```python
df_principals
```

```python

df_title_crew = pd.read_csv('C:/Users/nagor/Documents/Thebridge/Machine_Learning/imdb/datasets_brutos/title.crew.tsv.gz', 
                 sep='\t',          # separador de tabulaciones
                 compression='gzip',
                 low_memory=False)  #
```

```python
df_title_crew
```

```python
#Filtrar ratings relevantes
ratings_filtered = df_titulo_rating[df_titulo_rating['numVotes'] > 100]

#Unir con title.basics
merged_df = pd.merge(ratings_filtered, df_titulo_basic, on='tconst', how='inner')

#  Unir con crew
merged_df = pd.merge(merged_df, df_title_crew, on='tconst', how='left')

#Principals ‚Äî solo actores principales y no m√°s de 1 por t√≠tulo
principals_filtered = df_principals[df_principals['category'] == 'actor']
principals_top = principals_filtered.sort_values(by='ordering').drop_duplicates('tconst')

# Merge final con actor principal
merged_df = pd.merge(merged_df, principals_top[['tconst', 'nconst', 'characters']], on='tconst', how='left')

# Reemplazar '\N' por NaN
merged_df.replace('\\N', pd.NA, inplace=True)


# Eliminar filas con valores faltantes en columnas clave
columns_to_check = ['directors', 'writers', 'nconst', 'characters']
merged_df = merged_df.dropna(subset=columns_to_check)

# orden
merged_df.reset_index(drop=True, inplace=True)

final_df = merged_df
```

```python

# Eliminar columna isAdult
final_df.drop(columns=['isAdult'], inplace=True)

# Eliminar originalTitle (nos quedamos solo con primaryTitle)
final_df.drop(columns=['originalTitle'], inplace=True)
```

```python

# Eliminar '\N' (faltantes) de todo el dataframe
final_df.replace('\\N', pd.NA, inplace=True)



# Merge para agregar nombre del actor principal
final_df = pd.merge(final_df, df_actor[['nconst', 'primaryName']], on='nconst', how='left')
```

```python
# Eliminar columnas innecesarias
final_df.drop(columns=['tconst', 'writers', 'characters', 'endYear'], inplace=True)
```
